# AI_Assistant_Resume

```python
# Import the important libraries
import pdfminer
from pdfminer.high_level import extract_text
import openai
```

```python
# Extract the text from Resume
def extract_text_from_pdf(pdf_path):
    return extract_text(pdf_path)

# Replace 'your_pdf_file.pdf' with the path to your PDF file
pdf_path = 'Amir_DL.pdf'
extracted_text = extract_text_from_pdf(pdf_path)

print(extracted_text)
```

```python
# Replace 'YOUR_API_KEY' with your actual OpenAI API key
api_key = "sk-t........."

# Set the API key
openai.api_key = api_key

# Build the AI Assistant which is trained on my resume
def get_answer(question, context):
    prompt = f"Question: {question}\nContext: {context}\nAnswer:"
    response = openai.Completion.create(
        engine="text-davinci-003",  # or use "text-davinci-002" for GPT-3.5 Turbo
        prompt=prompt,
        max_tokens=100,
        n=1,
        stop=None
    )

    return response.choices[0].text.strip()

# Chatbot loop conditions
while True:
    user_question = input("User: ")
    if user_question.lower() == 'exit':
        print(">> : Thank You! It's my pleasure to chat with you")
        break

    answer = get_answer(user_question, extracted_text)
    print(f">> : {answer}\n")
```
